## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    palette='deep'
)
plt.title('Bill Length vs Bill Depth by Penguin Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.show()
```

**Your interpretation (2–3 sentences):**

Gentoo penguins generally have longer but shallower bills while Adelie penguins have shorter, deeper bills. Chinstrap penguins fall between the two. The color grouping makes it clear that each species occupies a distinct region of the plot.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
plt.figure(figsize=(8,6))
sns.histplot(penguins['bill_length_mm'], bins=10, color='skyblue')
plt.title('Bill Length Distribution (10 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(penguins['bill_length_mm'], bins=30, color='salmon')
plt.title('Bill Length Distribution (30 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

**Your comparison (1–2 sentences):**

With 10 bins, the histogram looks smoother and shows a general range of bill lengths. With 30 bins, the distribution looks more detailed which reveals subtle peaks that show differences among species.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
plt.figure(figsize=(8,6))
sns.boxplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    palette='pastel'
)
plt.title('Body Mass by Penguin Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```

**Your summary (1–2 sentences):**

Gentoo penguins are the heaviest on average while Adelie penguins are the lightest. Chinstrap penguins fall between the 2 with some variation compared to the others.
